```python
import pandas as pd
import altair as alt
```

## Data Summary

```python
tests = pd.read_csv('StudentsPerformance.csv')
tests['test score'] = tests.mean(axis=1)
print(tests.head())
```

## Final Visualization

```python
math = alt.Chart(tests).mark_circle().encode(
    x = 'math score',
    y = 'writing score',
    color = alt.Color('test preparation course', scale=alt.Scale()),
    tooltip = ('parental level of education', 'race/ethnicity', 'gender'),
    size=alt.Size('parental level of education',
        scale=alt.Scale(range=(20,100))
    ),
).properties(
    title = 'Math vs. Writing Score'
).interactive()

reading = alt.Chart(tests).mark_circle().encode(
    x = 'reading score',
    y = 'writing score',
    color = alt.Color('test preparation course', scale=alt.Scale()),
    tooltip = ('parental level of education', 'race/ethnicity', 'gender'),
    size=alt.Size('parental level of education',
        scale=alt.Scale(range=(20,100))
    ),
).properties(
    title = 'Reading vs. Writing Score'
).interactive() 

writing = alt.Chart(tests).mark_circle().encode(
    x = 'math score',
    y = 'reading score',
    color = alt.Color('test preparation course', scale=alt.Scale()),
    tooltip = ('parental level of education', 'race/ethnicity', 'gender'),
    size=alt.Size('parental level of education',
        scale=alt.Scale(range=(20,100))
    ),
).properties(
    title = 'Math vs. Reading Score'
).interactive()

plt = math | reading | writing


math | reading | writing
```

```python
plt.save('viz.html')
```
